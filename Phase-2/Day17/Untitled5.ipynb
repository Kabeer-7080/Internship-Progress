```python
# Import Required Libraries

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Iris Dataset

iris = sns.load_dataset("iris")

# Display the first 5 rows

iris.head()

# Line Plot of Sepal Length

plt.figure(figsize=(8,5))

plt.plot(iris.index, iris["sepal_length"], marker='o')

plt.title("Line Plot of Sepal Length")
plt.xlabel("Sample Index")
plt.ylabel("Sepal Length (cm)")

plt.grid(True)
plt.show()
# Bar Chart

plt.figure(figsize=(8,5))

sns.barplot(
    data=iris,
    x="species",
    y="sepal_length",
    hue="species",
    palette="Set2",
    legend=False
)

plt.title("Average Sepal Length by Species")
plt.xlabel("Species")
plt.ylabel("Average Sepal Length (cm)")

plt.show()
# Histogram

plt.figure(figsize=(8,5))

sns.histplot(
    data=iris,
    x="sepal_length",
    bins=10,
    color="skyblue",
    kde=True
)

plt.title("Distribution of Sepal Length")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Frequency")

plt.show()
# Scatter Plot

plt.figure(figsize=(8,5))

sns.scatterplot(
    data=iris,
    x="sepal_length",
    y="petal_length",
    hue="species",
    palette="Set2",
    s=100
)

plt.title("Sepal Length vs Petal Length")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Petal Length (cm)")

plt.show()
# Box Plot

plt.figure(figsize=(8,5))

sns.boxplot(
    data=iris,
    x="species",
    y="sepal_length",
    hue="species",
    palette="Set2",
    legend=False
)

plt.title("Sepal Length Distribution by Species")
plt.xlabel("Species")
plt.ylabel("Sepal Length (cm)")

plt.show()
# Count Plot

plt.figure(figsize=(8,5))

sns.countplot(
    data=iris,
    x="species",
    hue="species",
    palette="Set2"
)

plt.title("Count of Samples by Species")
plt.xlabel("Species")
plt.ylabel("Count")

plt.show()
# Pair Plot

sns.pairplot(
    data=iris,
    hue="species",
    palette="Set2"
)

plt.show()
```
