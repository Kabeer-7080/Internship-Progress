# Machine Learning Dataset Research

## Dataset 1: Online Shoppers Purchasing Intention Dataset

**Dataset Source:** https://archive.ics.uci.edu/dataset/468/online+shoppers+purchasing+intention+dataset

**Problem Statement:** Predict whether an online visitor will make a purchase during a browsing session.

- **Machine Learning Problem Type:** Classification
- **Number of Features:** 17 input features (18 columns including target)
- **Target Variable:** Revenue

### Reason for Selecting

- Real-world e-commerce dataset
- Beginner-friendly
- Suitable for comparing multiple classification algorithms

## Dataset 2: Customer Shopping Trends Dataset

**Dataset Source:** https://www.kaggle.com/datasets/iamsouravbanerjee/customer-shopping-trends-dataset

**Problem Statement:** Analyze customer shopping behavior and purchasing patterns based on demographic and transaction information.

- **Machine Learning Problem Type:** Classification
- **Number of Features:** 18 input features (19 columns including target)
- **Target Variable:** Purchased Item

### Reason for Selecting

- Real-world customer shopping dataset
- Useful for customer behavior analysis
- Suitable for recommendation and classification tasks

## Dataset 3: E-Commerce Shipping Data Dataset

**Dataset Source:** https://www.kaggle.com/datasets/prachi13/customer-analytics

**Problem Statement:** Predict whether an e-commerce order will be delivered on time based on customer, product, and shipping information.

- **Machine Learning Problem Type:** Classification
- **Number of Features:** 11 input features (12 columns including target)
- **Target Variable:** Reached.on.Time_Y.N

### Reason for Selecting

- Real-world e-commerce logistics dataset
- Useful for delivery time prediction
- Suitable for comparing multiple classification algorithms

# Final Selected Dataset

## Online Shoppers Purchasing Intention Dataset

### Justification

The Online Shoppers Purchasing Intention Dataset was selected because it is a real-world e-commerce dataset, well-structured, and widely used for binary classification. It allows easy implementation and comparison of machine learning algorithms such as Logistic Regression, K-Nearest Neighbors (KNN), Decision Tree, Random Forest, and Support Vector Machine (SVM).
